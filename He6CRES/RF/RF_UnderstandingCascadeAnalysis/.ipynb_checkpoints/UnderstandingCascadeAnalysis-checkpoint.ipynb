# Toy Model: 

* We can understand most of our system by modeling the Terminator, Waveguides, and two LNAs. 
* To do this we think of our system as being composed of: 
    * Some noise coming in (N_i): This is noise sourced from the terminator. The terminator sources noise proportionally to it's temperature: (N_i = kBT_Term, T_0 = T_Term)
    * Three components with gain (3 boxes). WG assembly, LNA1, LNA2. But how do you get the equivalent noise temps of these, T_e?
        * WG: Use eq. 10.15
        * LNAs: Look at datasheets.
* You take these parameters and plug them into 10.22. 


## Modeling the Equivalent Noise Temp of 6He CRES

```python
# Functions: 

# Converts a value in dB to linear units
dBToLinear = lambda x: 10**(x/10) 
# 
T_e_tot = lambda T_Term, T_1, T_e_2, T_e_3, G_1, G_2: T_Term + T_1*(dBToLinear(G_1)**-1 -1) + T_e_2/(dBToLinear(G_1)) + T_e_3/(dBToLinear(G_1 + G_2)) # eq. 10.22
```

# Legend: 

| Parameter  | Denoted  |
|---|---|
| Terminator Temperature  |  T_Term |
| Waveguide Temperature  | T_1  |
| Equivalent Noise Temp of Stage 1 LNA  | T_e_2  |
| Equivalent Noise Temp of Stage 2 LNA  |  T_e_3 |
|  Gain of Waveguides  |  G_1 |
|  Gain of LNA1  |  G_2 |
|  Gain of LNA2  |  G_3 |

## Formulas Used:  

```python
%%latex
$T_{Tot} = T_{term} + T_1(\frac{1}{G_1}-1)+ \frac{T_{e2}}{G_1} + \frac{T_{e3}}{G_1 G_2} $ (based on eq 10.22)
```

```python
%%latex
$N_{Density}/k_B =  G_1 G_2 G_3 T_{Tot}  $ (based on eq 10.19)
```

```python
NoiseTempTot = interactive(T_e_tot, T_Term = (13,43),T_1=(50,150), T_e_2 = (6,8), T_e_3 = (8,10),G_1=(-4,0), G_2=(31,33))
display(NoiseTempTot)
```

```python
G_3 = 31
k = 1.380649* 10**-23  # (J/K)G_2=(31,33)
B = 170 *10**3         # (Hz)

print("Noise Temperature of System:" , NoiseTempTot.result, "K")
print("Noise Density of (abbreviated) System:" , (NoiseTempTot.kwargs['G_1'] + NoiseTempTot.kwargs['G_2'] + G_3)+ 10*np.log10(NoiseTempTot.result*k*1000), "dBm/Hz")
```

```python
# Plot of ND vs G_1
G_1 = np.arange(-4,.1,.1)
k_B = 1.38e-23
# T_Term, T_1, T_e_2, T_e_3, G_1, G_2
plt.figure(1)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.xlabel('Gain of WG (dB)', fontsize = 10)
plt.ylabel('Noise Temperature (K)', fontsize = 10)
plt.title("Total Noise Temperature vs Gain of Waveguides")

plt.plot(G_1, T_e_tot(NoiseTempTot.kwargs['T_Term'], NoiseTempTot.kwargs['T_1'], NoiseTempTot.kwargs['T_e_2'], NoiseTempTot.kwargs['T_e_3'], G_1,NoiseTempTot.kwargs['G_2'] ) )
plt.show()


plt.figure(1)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.xlabel('Gain of WG (dB)', fontsize = 10)
plt.ylabel('Noise Density (dBm/Hz)', fontsize = 10)
plt.title("Total Noise Density vs Gain of Waveguides")

plt.plot(G_1, (G_1+NoiseTempTot.kwargs['G_2']+G_3+10*np.log10((1000*k_B*T_e_tot(NoiseTempTot.kwargs['T_Term'], NoiseTempTot.kwargs['T_1'], NoiseTempTot.kwargs['T_e_2'], NoiseTempTot.kwargs['T_e_3'], G_1,NoiseTempTot.kwargs['G_2'] ))) ))
plt.show()
```
