```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-2,2,20),np.linspace(-1,1,20))
pi = np.pi

# Plot for a surface current J that results in only TE_11 (m=n=1) excitation: 
# a = 2, b = 1, A = 1 

u = np.cos(pi*x/2)*np.sin(pi*y)
v = .5*np.sin(pi*x/2)*np.cos(pi*y)

plt.quiver(x,y,u,v)
plt.show()
```

```python
# Plot for a surface current J that results in only TE_10 (m=1, n=0) excitation: 
# a = 2, b = 1, A = 1 

u = 0
v = .5*np.sin(pi*x/2)

plt.quiver(x,y,u,v)
plt.show()
```

But remember J is a phasor here.
